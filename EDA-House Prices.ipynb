# checking relationship between sales price and  missing value

```python
for feature in null_in_feature:
    data = df.copy()
    data[feature] = np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot(kind='bar')
    plt.title(feature)
    plt.show()
```

# let's check discreat value

```python
for i in discreat_value:
    dataset = data.copy()
    sns.set_theme(style="whitegrid")
    fig = plt.gcf()
    fig.set_size_inches(12, 8)
    sns.catplot(data=dataset,x=i,y='SalePrice',kind='bar',ci=None)
```

```python
contunes_varable =[x for x in numicarl_value if x not in discreat_value and x not in year_var + ['Id'] ]
```

```python
len(contunes_varable)
```

```python
for i in contunes_varable:
    dataest = data.copy()
    sns.displot(dataset, x=dataest[i])
```

```python
#appling log tranformation 
```

```python
for x in contunes_varable:
    data_set = data.copy()
    if 0 in data_set[x].unique():
        pass
    else:
        data_set[x] = np.log(data_set[x])
        data_set['SalePrice']=np.log(data_set['SalePrice'])
        sns.relplot(x= data_set[x], y="SalePrice", data=data_set)
```
